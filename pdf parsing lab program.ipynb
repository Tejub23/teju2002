```python
import fitz 
def extract_images_from_page(pdf_file, page_number):
    doc = fitz.open(pdf_file)
    images = []
    if page_number < 1 or page_number > len(doc):
        print(f"Page number {page_number} is out of range.")
        return images
    page = doc.load_page(page_number - 1)  
    for img_index, img in enumerate(page.get_images(full=True)):
        xref = img[0]
        base_image = doc.extract_image(xref)
        images.append({
            'page_number': page_number,
            'image_index': img_index,
            'image': base_image['image']
        })

    doc.close()
    return images

pdf_file = 'TOOTH-FAIRY.pdf'
user_input_page_number = int(input("Enter the page to exatrct the image: ")) 
extracted_images = extract_images_from_page(pdf_file, user_input_page_number)
for img_info in extracted_images:
    image = img_info['image']
    image_file_name = f"page {img_info['page_number']}.png"
    with open(image_file_name, 'wb') as f:
        f.write(image)

    print(f"Image extracted from page {img_info['page_number']} saved to {image_file_name}")
```

```python
import fitz  

def extract_links_from_pdf(pdf_file):
    doc = fitz.open(pdf_file)
    links = []

    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        page_links = page.get_links()
        
        for link in page_links:
            if 'uri' in link:
                link_info = {
                    'page_number': page_num + 1,
                    'uri': link['uri']
                }
                if 'rect' in link:
                    link_info['rectangle'] = link['rect']
                else:
                    link_info['rectangle'] = None 
                links.append(link_info)

    doc.close()
    return links
pdf_file = 'TOOTH-FAIRY.pdf'
extracted_links = extract_links_from_pdf(pdf_file)
for link in extracted_links:
    print(f"Page {link['page_number']}: {link['uri']} at {link['rectangle']}")
```

```python
import fitz 
def merge_pdfs(pdf_file1, pdf_file2, output_pdf):
    doc1 = fitz.open(pdf_file1)
    doc2 = fitz.open(pdf_file2)
    merged_pdf = fitz.open()
    for page_num in range(len(doc1)):
        merged_pdf.insert_pdf(doc1, from_page=page_num, to_page=page_num)
    for page_num in range(len(doc2)):
        merged_pdf.insert_pdf(doc2, from_page=page_num, to_page=page_num)
    merged_pdf.save(output_pdf)
    doc1.close()
    doc2.close()
    merged_pdf.close()
pdf_file1 = 'TOOTH-FAIRY.pdf'
pdf_file2 = 'Beauty-and-the-Beast.pdf'
output_pdf = 'mergedpdf.pdf'

merge_pdfs(pdf_file1, pdf_file2, output_pdf)

print(f"Merged PDF saved as {output_pdf}")
```

```python
import pdfplumber
import csv

def extract_tables_from_pdf(pdf_file, output_csv=None):
    tables = []

    # Open the PDF file
    with pdfplumber.open(pdf_file) as pdf:
        # Iterate over each page
        for page in pdf.pages:
            # Extract tables from the page
            extracted_tables = page.extract_tables()
            
            for table in extracted_tables:
                for row in table:
                    # Combine multi-line cells into single line cells
                    row = [cell.replace('\n', ' ') if cell else '' for cell in row]
                    tables.append(row)
    
    # If output_csv is provided, save the tables to a CSV file
    if output_csv:
        with open(output_csv, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['Section', 'Function'])  # Add header
            for row in tables:
                writer.writerow(row)
    
    return tables

# Example usage:
pdf_file = 'learningguide.pdf'
output_csv = 'extracted_tables.csv'
extracted_tables = extract_tables_from_pdf(pdf_file, output_csv)

# Print the extracted tables in a readable format
for row in extracted_tables:
    print(row)

print(f"Tables extracted and saved to {output_csv}")
```

```python
import fitz 
def rotate_page(pdf_file, page_number, rotation_angle, output_pdf):
    doc = fitz.open(pdf_file)
    page = doc.load_page(page_number - 1)
    page.set_rotation(rotation_angle)
    doc.save(output_pdf)
    doc.close()
pdf_file = 'TOOTH-FAIRY.pdf'
page_number = int(input("Enter the page number"))
rotation_angle = 90 
output_pdf = 'rotate.pdf'

rotate_page(pdf_file, page_number, rotation_angle, output_pdf)

print(f"Page {page_number} rotated by {rotation_angle} degrees and saved to {output_pdf}")
```
